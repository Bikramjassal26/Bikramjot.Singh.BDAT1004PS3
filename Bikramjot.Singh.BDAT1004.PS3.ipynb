# Question 6

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
US_Mar_Div= pd.read_csv('us-marriages-divorces-1867-2014.csv')
US_Mar_Div.head()
```

```python
years = US_Mar_Div['Year']
marriages = US_Mar_Div['Marriages_per_1000']
divorces = US_Mar_Div['Divorces_per_1000']
US_Mar_Div = plt.figure(figsize=(20,16))
US_Mar_Div = plt.plot(years, marriages, label='Marriages', linewidth=4)
US_Mar_Div = plt.plot(years, divorces, label='Divorces', linewidth=4)
US_Mar_Div = plt.title("Number of marriages and divorces per capita between 1867 and 2014",fontsize=24)
US_Mar_Div = plt.xlabel("Years",fontsize=30)
US_Mar_Div = plt.legend(fontsize = 24, loc = "upper left")
US_Mar_Div = plt.ylabel("Marriages",fontsize=30)
US_Mar_Div
```

# Question 7

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
Bar_chart_data = pd.read_csv('us-marriages-divorces-1867-2014.csv')
Bar_chart_data.head()
```

```python
bar_chart=Bar_chart_data.loc[Bar_chart_data.Year.isin([1900,1950,2000]),['Year','Marriages_per_1000','Divorces_per_1000']]
bar_chart
```

```python
bar_chart = bar_chart[bar_chart['Year'].apply(lambda x: x in [1900, 1950, 2000])]
years = bar_chart['Year']
marriages = bar_chart['Marriages_per_1000']
divorces = bar_chart['Divorces_per_1000']
bar_chart = plt.figure(figsize= (20,16))
bar_chart = plt.bar(years, marriages, label ='Marriages', width=16, color='red')
bar_chart = plt.bar(years, divorces, label = 'Divorces', width = 16)
bar_chart = plt.title("Number of marriages and divorces per capita between 1990,1950 and 2000", fontsize=24)
bar_chart = plt.xlabel("Years", fontsize=30)
bar_chart = plt.legend(fontsize = 20, loc = "upper left")

bar_chart
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
actor_kills = pd.read_csv('actor_kill_counts.csv')
actor_kills
```

```python
actors = actor_kills['Actor']
killCount = actor_kills['Count']
actor_kills = plt.figure(figsize=(16,16), facecolor='lightgray')
actor_kills = plt.barh(actors, killCount, label='Count', color='red')
actor_kills = plt.title("Deadliest actors in Hollywood", fontsize=28)
actor_kills = plt.xlabel("Kill count", fontsize=24)
actor_kills = plt.ylabel("Actors", fontsize = 24)
actor_kills = plt.legend(fontsize = 24, loc = "upper right")
actor_kills
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plot
```

```python
Scatterplot = pd.read_csv("arcade-revenue-vs-cs-doctorates.csv")
Scatterplot.rename(columns = {'Total Arcade Revenue (billions)':'Revenue', 'Computer Science Doctorates Awarded (US)':'Awards'}, inplace = True)
Scatterplot
```

```python
a = Scatterplot.groupby('Year')
for name, group in a:
    plot.plot(group.Revenue, group.Awards, marker='o', linestyle='', markersize=20, label=name)
plot.legend()
plot.xlabel("Total revenue earned by Arcades")
plot.ylabel("Computer Science PhD Awards")
plot.figure(figsize=(20,20))
```
